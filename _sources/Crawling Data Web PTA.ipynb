# Crawling Data

<p>Code di bawah ini digunakan untuk mengcrawling data dari sebuah website yang nantinya akan diolah menggunakan stemming, tf-idf dan lain sebagainya</p>

```python
import scrapy


class QuotesSpider(scrapy.Spider):
    name = "quotes"

    def start_requests(self):
        start_urls = [
            'https://pta.trunojoyo.ac.id/welcome/index/12',
            'https://pta.trunojoyo.ac.id/welcome/index/13',
            'https://pta.trunojoyo.ac.id/welcome/index/14',
            'https://pta.trunojoyo.ac.id/welcome/index/15',
            'https://pta.trunojoyo.ac.id/welcome/index/16',
        ]

        for url in start_urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for item in response.css('#content_journal > ul > li'):
            yield {
                'Link': item.css(f'div:nth-child(3) > div:nth-child(5) > a::attr(href)').get(),
            }
```
